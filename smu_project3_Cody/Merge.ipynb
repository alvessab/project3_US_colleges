```python
filepath = "../smu_project3_Cody/csv/degrees-that-pay-back.csv"
filepath2 = "../smu_project3_Cody/csv/IPED_college_information.csv"
filepath3 = "../smu_project3_Cody/csv/salaries-by-college-type.csv"
filepath4 = "../smu_project3_Cody/csv/salaries-by-region.csv"

df_deg = pd.read_csv(filepath)
df_iped = pd.read_csv(filepath2)
df_type = pd.read_csv(filepath3)
df_region = pd.read_csv(filepath4)
```

```python
clean_df = merge_df2.filter(['School Name', 'Region', 'Starting Median Salary_x',
       'Mid-Career Median Salary_x', 'Mid-Career 10th Percentile Salary_x',
       'Mid-Career 25th Percentile Salary_x',
       'Mid-Career 75th Percentile Salary_x',
       'Mid-Career 90th Percentile Salary_x', 'STABBR', 'CBSA', 'LONGITUD', 'LATITUDE', "School Type"], axis =1)
clean_df.head()
```

```python
clean_df.info()
```

```python
clean_df.rename(columns={'Starting Median Salary_x': "starting_median_salary", 'Mid-Career Median Salary_x':'career_median_salary',
                        'LONGITUD':'LONGITUDE'}, inplace = True)
clean_df.head()
```

```python
clean_df.to_csv("merged.csv", index=False, header=True)
```

```python
clean_df['starting_median_salary'] = clean_df['starting_median_salary'].replace('$', '')
clean_df.head()
```

```python
# https://stackoverflow.com/questions/31521526/convert-currency-to-float-and-parentheses-indicate-negative-amounts
clean_df[['starting_median_salary']] = clean_df[['starting_median_salary']].replace('[\$,]','',regex=True).astype(float)
clean_df[['career_median_salary']] = clean_df[['career_median_salary']].replace('[\$,]','',regex=True).astype(float)
clean_df.head()
```

```python
grouped = clean_df.groupby("Region")
grouped.mean()
```
